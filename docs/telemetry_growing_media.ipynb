The "Growing Media" pillar focuses on the substrate or medium in which the cannabis plants grow. This can include soil, coco coir, hydroponic systems, aeroponic systems, and more. The quality and condition of the growing medium directly impact plant health, nutrient uptake, water retention, and root growth.

For this pillar, let's consider the following metrics:

pH Level: This indicates the acidity or alkalinity of the growing medium. Most cannabis plants prefer a slightly acidic pH.
Moisture Content: This represents the amount of water retained in the medium, critical for plant hydration and nutrient uptake.
Oxygen Level: In hydroponic systems, the oxygen level in the growing medium is crucial for root health.
# Step 1: Define the Healthy Ranges
pH Level: 6.0-7.0
Moisture Content: 50-70% for soil (this can vary depending on the specific medium)
Oxygen Level (for hydroponic systems): 5-8 mg/L
# Step 2: Create the Simulator

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(mean, std_dev, size=100):
    """Generate random data based on mean and standard deviation."""
    return np.random.normal(mean, std_dev, size)

# Define the metrics for Growing Media
# pH Level
ph_healthy_mean = 6.5
ph_healthy_std_dev = 0.2
ph_unhealthy_data = np.array([5.8, 5.7, 7.2, 7.3])
ph_data = np.insert(generate_data(ph_healthy_mean, ph_healthy_std_dev), [20, 40, 60, 80], ph_unhealthy_data)

# Moisture Content
moisture_healthy_mean = 60
moisture_healthy_std_dev = 5
moisture_unhealthy_data = np.array([45, 44, 75, 78])
moisture_data = np.insert(generate_data(moisture_healthy_mean, moisture_healthy_std_dev), [20, 40, 60, 80], moisture_unhealthy_data)

# Oxygen Level
oxygen_healthy_mean = 6.5
oxygen_healthy_std_dev = 0.5
oxygen_unhealthy_data = np.array([4.5, 4.0, 8.5, 9.0])
oxygen_data = np.insert(generate_data(oxygen_healthy_mean, oxygen_healthy_std_dev), [20, 40, 60, 80], oxygen_unhealthy_data)

# Plotting the data
plt.figure(figsize=(15, 5))

# Plotting pH Level
plt.subplot(1, 3, 1)
plt.plot(ph_data)
plt.axhline(y=6.0, color='r', linestyle='--')
plt.axhline(y=7.0, color='r', linestyle='--')
plt.title("pH Level Telemetry")
plt.ylabel("pH Level")
plt.xlabel("Time")

# Plotting Moisture Content
plt.subplot(1, 3, 2)
plt.plot(moisture_data)
plt.axhline(y=50, color='r', linestyle='--')
plt.axhline(y=70, color='r', linestyle='--')
plt.title("Moisture Content Telemetry")
plt.ylabel("Moisture Content (%)")
plt.xlabel("Time")

# Plotting Oxygen Level
plt.subplot(1, 3, 3)
plt.plot(oxygen_data)
plt.axhline(y=5, color='r', linestyle='--')
plt.axhline(y=8, color='r', linestyle='--')
plt.title("Oxygen Level Telemetry")
plt.ylabel("Oxygen Level (mg/L)")
plt.xlabel("Time")

# Adjust the layout
plt.tight_layout()
plt.show()
```

# Step 3: Plotting for Jupyter Notebooks
The plots above represent the telemetry data for the "Growing Media" pillar:

pH Level: The ideal range is between 6.0 and 7.0, with the red dashed lines indicating the boundaries. Values outside of this range can affect the plant's ability to absorb nutrients.
Moisture Content: For soil-based media, the moisture content should be between 50% and 70%. Over-watering or under-watering can stress the plant and affect its growth.
Oxygen Level: For hydroponic systems, an oxygen level between 5-8 mg/L is considered healthy. Too low oxygen can lead to root rot, while too much can dry out the roots.
# Step 4: Explanation & Analysis
pH Level:
Importance: The pH level of the growing medium affects the plant's ability to absorb nutrients. At extreme pH values, certain nutrients become unavailable to the plant, potentially leading to nutrient deficiencies.
Impact of Deviations: A pH level that's too high or too low can lead to nutrient lockout, where the plant can't absorb the necessary nutrients, even if they're present in the medium.
Corrective Actions: Regularly monitor the pH and adjust using pH up or down solutions. Regularly flush the growing medium with pH-balanced water to prevent buildup of salts or other compounds.
Moisture Content:
Importance: Maintaining the right moisture level ensures the plant roots get enough water without being waterlogged.
Impact of Deviations: Over-watering can lead to root rot, while under-watering can stress the plant and reduce growth.
Corrective Actions: Use moisture meters to regularly check the moisture content. Ensure proper drainage and avoid letting plants sit in stagnant water.
Oxygen Level (for hydroponic systems):
Importance: Oxygen is vital for root respiration. Without adequate oxygen, roots can't effectively absorb nutrients and water.
Impact of Deviations: Low oxygen levels can lead to root rot and slow growth. Excessively high oxygen levels can dry out the roots.
Corrective Actions: In hydroponic systems, ensure water is well-aerated using air stones or diffusers. Regularly check and adjust the oxygen level as needed.
By understanding these metrics and their significance, cultivators can make informed decisions to optimize their growing conditions, leading to healthier plants and better yields.
