# Step 2: Create the Simulator
We'll use the function you provided to generate data and create a simulated data stream for temperature.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_data(mean, std_dev, size=100):
    return np.random.normal(mean, std_dev, size)

# Parameters for Temperature
temp_healthy_mean = 25  # Central value of the healthy range
temp_healthy_std_dev = 2  # Adjust for desired fluctuation

# Generate healthy data
temp_healthy_data = generate_data(temp_healthy_mean, temp_healthy_std_dev)

# Introduce some unhealthy data
temp_unhealthy_data = np.array([15, 35, 36, 17, 18])
temp_data = np.insert(temp_healthy_data, [20, 40, 60, 80, 90], temp_unhealthy_data)
```

```python
plt.figure(figsize=(12, 6))
plt.fill_between(range(len(temp_data)), 20, 30, color='green', alpha=0.1, label="Healthy Range")
plt.plot(temp_data, label="Temperature")
plt.axhline(y=20, color='r', linestyle='--')
plt.axhline(y=30, color='r', linestyle='--')
plt.title("Temperature Telemetry")
plt.ylabel("Temperature (Â°C)")
plt.xlabel("Time")
plt.legend()
plt.show()
```

# Step 4: Explanation & Analysis
Temperature:

Importance: Temperature plays a crucial role in cannabis plant health. It affects metabolic rates, growth speed, and overall health. Proper temperatures ensure that the plant can effectively undergo photosynthesis and other vital processes.
Impact of deviations:
Below the healthy range: Slows metabolic rates, stunting growth and potentially causing the plant to go dormant. Cold temperatures can also damage or kill the plant.
Above the healthy range: Increases metabolic rates to potentially harmful levels, can cause heat stress, wilting, and reduced growth. Prolonged exposure to high temperatures can severely harm or kill the plant.
Potential reasons for the deviation:
Low Temperature: Lack of adequate heating, external weather conditions, malfunctioning equipment.
High Temperature: Excessive lighting, lack of ventilation, malfunctioning cooling systems.
Corrective actions: Monitor the growing environment regularly. Use heaters or cooling systems as needed. Ensure proper ventilation and adjust lighting if necessary.
