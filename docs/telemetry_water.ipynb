# Water

# Step 1: Define the Healthy Ranges
For the "Water Quality", the following parameters are typically considered crucial:

pH: 5.5-6.5 (This range ensures optimal nutrient uptake for cannabis plants)
Electrical Conductivity (EC): 1.2-2.2 mS/cm (A measure of the nutrient concentration in the water)
Total Dissolved Solids (TDS): 800-1500 ppm (This represents the total amount of minerals, salts, and metals dissolved in water)
Water Temperature: 18-22Â°C (For optimal root health and nutrient uptake)
# Step 2: Create the Simulator
Let's generate simulated time-series data for the Water Quality pillar:

We'll begin with pH:

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(mean, std_dev, size=100):
    return np.random.normal(mean, std_dev, size)
# Parameters for pH
pH_healthy_mean = 6  # Central value of the healthy range
pH_healthy_std_dev = 0.2  # Adjust for desired fluctuation

# Generate healthy data for pH
pH_healthy_data = generate_data(pH_healthy_mean, pH_healthy_std_dev)

# Introduce some unhealthy data for pH
pH_unhealthy_data = np.array([5, 7, 7.2, 5.1, 4.8])
pH_data = np.insert(pH_healthy_data, [20, 40, 60, 80, 90], pH_unhealthy_data)

plt.figure(figsize=(12, 6))
plt.fill_between(range(len(pH_data)), 5.5, 6.5, color='green', alpha=0.1, label="Healthy Range")
plt.plot(pH_data, label="Water pH")
plt.axhline(y=5.5, color='r', linestyle='--')
plt.axhline(y=6.5, color='r', linestyle='--')
plt.title("Water pH Telemetry")
plt.ylabel("pH")
plt.xlabel("Time")
plt.legend()
plt.show()
```

This plot gives you a visual representation of the simulated pH values over time, with the green shaded region indicating the healthy range.

# Step 4: Explanation & Analysis
For the pH parameter:

Importance: pH indicates the acidity or alkalinity of the water. It's crucial for determining how well plants can take up nutrients available in the water.
Impact of deviations:
Below the healthy range: Nutrient absorption can be inhibited, leading to deficiencies.
Above the healthy range: Certain nutrients become unavailable, leading to nutrient lockout.
Potential reasons for the deviation:
Low or high pH: Could be due to the source water, or the nutrients and additives used.
Corrective actions: Adjust the pH using pH up or down solutions.
