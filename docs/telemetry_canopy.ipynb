# Canopy

# Step 1: Define the Healthy Ranges
For our Canopy simulation, we'll focus on:

Height: This can represent the growth of the plant. A continuous increase in height can be a sign of a healthy plant, but too rapid or too slow growth might be problematic.
Healthy Range: 50-200 cm
Density: Represents how closely packed the plant leaves and branches are. Too dense might mean poor light penetration, and too sparse might mean poor health.
Healthy Range: 0.5-0.9 (arbitrary units)
Leaf Area Index (LAI): Represents the leaf area per unit ground area. It's a good indicator of plant health and its potential photosynthesis capacity.
Healthy Range: 1-5
## Step 2: Create the Simulator
We will extend our IoTDevice class to simulate canopy telemetry. The class will be named CanopyMonitor.

First, let's define the CanopyMonitor class and simulate the Height telemetry.

```python
import numpy as np
import matplotlib.pyplot as plt

# Base IoTDevice class as defined previously
class IoTDevice:
    def __init__(self, device_id, location):
        self.device_id = device_id
        self.location = location
        self.telemetry_data = {}
        self.status = "offline"

    def connect(self):
        self.status = "online"

    def disconnect(self):
        self.status = "offline"

    def send_telemetry(self, data):
        pass

    def simulate_telemetry(self):
        pass


# Extended class for Canopy Monitoring
class CanopyMonitor(IoTDevice):
    def simulate_telemetry(self):
        # Simulate Height
        height_healthy_mean = 125  # Central value of the healthy range
        height_healthy_std_dev = 30  # Adjust for desired fluctuation

        # Generate healthy data
        height_healthy_data = generate_data(height_healthy_mean, height_healthy_std_dev)

        # Introduce some unhealthy data
        height_unhealthy_data = np.array([30, 250, 260, 45, 40])
        height_data = np.insert(height_healthy_data, [20, 40, 60, 80, 90], height_unhealthy_data)
        
        return height_data


# Function to generate data
def generate_data(mean, std_dev, size=100):
    return np.random.normal(mean, std_dev, size)


# Simulate Height Telemetry
device = CanopyMonitor(device_id="001", location="Greenhouse A")
height_data = device.simulate_telemetry()

plt.figure(figsize=(10, 5))
plt.plot(height_data, label="Canopy Height")
plt.axhline(y=50, color='r', linestyle='--', label="Healthy Lower Bound")
plt.axhline(y=200, color='r', linestyle='--', label="Healthy Upper Bound")
plt.title("Canopy Height Telemetry")
plt.ylabel("Height (cm)")
plt.xlabel("Time")
plt.legend()
plt.show()
```

The plot above represents the simulated telemetry data for canopy height. The red dashed lines indicate the boundaries of the "healthy" range, while any values outside of this range are considered "unhealthy".

```python
# Extended class for Canopy Monitoring with added Density telemetry
class CanopyMonitorDensity(IoTDevice):
    def simulate_telemetry(self):
        # Simulate Density
        density_healthy_mean = 0.7  # Central value of the healthy range
        density_healthy_std_dev = 0.1  # Adjust for desired fluctuation

        # Generate healthy data
        density_healthy_data = generate_data(density_healthy_mean, density_healthy_std_dev)

        # Introduce some unhealthy data
        density_unhealthy_data = np.array([0.3, 1.2, 1.3, 0.2, 0.25])
        density_data = np.insert(density_healthy_data, [20, 40, 60, 80, 90], density_unhealthy_data)
        
        return density_data


# Simulate Density Telemetry
device_density = CanopyMonitorDensity(device_id="002", location="Greenhouse B")
density_data = device_density.simulate_telemetry()

plt.figure(figsize=(10, 5))
plt.plot(density_data, label="Canopy Density")
plt.axhline(y=0.5, color='r', linestyle='--', label="Healthy Lower Bound")
plt.axhline(y=0.9, color='r', linestyle='--', label="Healthy Upper Bound")
plt.title("Canopy Density Telemetry")
plt.ylabel("Density (arbitrary units)")
plt.xlabel("Time")
plt.legend()
plt.show()
```

The above plot showcases the simulated telemetry data for canopy density. The "healthy" range is represented by the area between the two red dashed lines. Values outside of this region would be considered "unhealthy" or non-optimal for the plant's growth.

```python
# Extended class for Canopy Monitoring with added Leaf Area Index (LAI) telemetry
class CanopyMonitorLAI(IoTDevice):
    def simulate_telemetry(self):
        # Simulate LAI
        lai_healthy_mean = 3  # Central value of the healthy range
        lai_healthy_std_dev = 0.5  # Adjust for desired fluctuation

        # Generate healthy data
        lai_healthy_data = generate_data(lai_healthy_mean, lai_healthy_std_dev)

        # Introduce some unhealthy data
        lai_unhealthy_data = np.array([1, 5, 5.5, 1.5, 0.8])
        lai_data = np.insert(lai_healthy_data, [20, 40, 60, 80, 90], lai_unhealthy_data)
        
        return lai_data


# Simulate LAI Telemetry
device_lai = CanopyMonitorLAI(device_id="003", location="Greenhouse C")
lai_data = device_lai.simulate_telemetry()

plt.figure(figsize=(10, 5))
plt.plot(lai_data, label="Leaf Area Index (LAI)")
plt.axhline(y=2, color='r', linestyle='--', label="Healthy Lower Bound")
plt.axhline(y=4, color='r', linestyle='--', label="Healthy Upper Bound")
plt.title("Leaf Area Index (LAI) Telemetry")
plt.ylabel("LAI")
plt.xlabel("Time")
plt.legend()
plt.show()
```

The plot above depicts the simulated telemetry data for the Leaf Area Index (LAI) of the canopy. The "healthy" range for LAI is typically between the values of 2 and 4 (represented by the area between the two red dashed lines). Values outside of this range may indicate issues with the plant canopy, such as reduced photosynthesis efficiency or other growth-related challenges.

To summarize:

Importance of LAI: The Leaf Area Index (LAI) is a measure of plant canopy density. It represents the total one-sided area of leaf tissue per unit ground surface area. LAI is an important parameter in agriculture and forestry, as it provides insight into the plant's photosynthesis capacity. A healthy LAI indicates optimal light interception by the plant, which in turn affects the overall plant growth and yield.

Impact of Deviations:

Low LAI: Indicates less leaf area, which might mean reduced photosynthesis and potential undergrowth of the plant. It can be caused by factors like pest infestations, diseases, nutrient deficiencies, or inadequate light.
High LAI: While having a high leaf area can be beneficial up to a certain extent, an overly dense canopy can hinder light penetration to the lower leaves, potentially causing them to become unproductive. It can also lead to issues like increased humidity within the canopy, fostering disease outbreaks.
Corrective Actions:

For Low LAI: Check for pests or diseases. Ensure the plant is receiving adequate nutrients and light. Adjust watering routines if needed.
For High LAI: Consider pruning or thinning practices to allow better light penetration. Ensure adequate airflow to prevent the buildup of humidity.
